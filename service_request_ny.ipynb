##### As we get the elapsed time between requrest created and closed request so we can store them from datatime datatype to integer type and can store in a column with hours.

```python
time_ct_hrs = service_req[['City','Complaint Type','Request_Closing_Time']]
time_ct_hrs.dropna(subset = ['City','Complaint Type','Request_Closing_Time'],inplace = True)
time_ct_hrs['Delta_in(hrs)'] = np.around((time_ct_hrs['Request_Closing_Time'].astype(np.int64)/(pow(10,9)*3600)),decimals = 2)
neg_time = time_ct_hrs[time_ct_hrs['Delta_in(hrs)']<0].sum()
print('Show the time as there is no negative time and also it is not possible(Time Created>Time Closed not possible)',neg_time)
```

##### Performing the One sample t -test

```python
print('T-statistics test',ttest_one_sample)
print('P-Value',np.around(p_val))
```

```python
if p_val<0.05:
    print('Value is less than P-values({})so reject the null hypothesis'.format(np.around(p_val)))
else:
    print('Value is more than P-values({})so fail to reject the null hypothesis'.format(np.around(p_val)))
```

##### Now performing the two sample ttest.

```python
### Getting the two samples from the population data.
sample1 = avg_response_time.sample(frac = 0.5)
sample1
```

```python
sample2 = avg_response_time.drop(sample1.index)
sample2
```

#### Thus how we get the two samples for perform the two sample t test.

```python
# Performing the two sample ttest.
# Getting the mean and std deviation of sample 1 and sample 2 for that.
sample1_mean,sample1_std = sample1['Delta_in(hrs)'].agg({'mean','std'})
sample1_mean,sample1_std
```

```python
sample2_mean,sample2_std = sample2['Delta_in(hrs)'].agg({'mean','std'})
sample2_mean,sample2_std
```

```python
# Now performing the t test for sample1 and sample2.
ttest_samp2 , pvalue = stat.ttest_ind(sample1['Delta_in(hrs)'],sample2['Delta_in(hrs)'])
```

```python
print('T-Statistics({})'.format(ttest_samp2))
```

```python
print('P - Value is ({})' .format(pvalue))
```

```python
if pvalue<0.05:
    print("P - Value is ({})if p value is less than 0.05 than reject the null hypothesis".format(np.around(pvalue),decimals = 2))
else:
    print("P - Value is ({}) which is greater than 0.05 than fail reject the null hypothesis or accept the null hypothesis".format(np.around(pvalue),decimals = 2))
```
